```python
# Predictive Imputation:

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Assuming df is your DataFrame and it's already cleaned

# Select features that you think may predict ConvertedComp
features = ['EdLevel', 'JobSat']  # Add more features as necessary
X = df[features]
y = df['ConvertedComp']

# Check for missing values
missing_count = df['ConvertedComp'].isna().sum()
print(f"Initial number of missing values in 'ConvertedComp': {missing_count}")

if missing_count > 0:
    # Create a Column Transformer
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore'), features)  # One-hot encode categorical features
        ],
        remainder='drop'  # Drop other columns
    )

    # Create a pipeline that first transforms the data, then fits the model
    model = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', LinearRegression())
    ])

    # Prepare your training data (excluding rows where ConvertedComp is NaN)
    train_data = df[df['ConvertedComp'].notna()]
    X_train = train_data[features]
    y_train = train_data['ConvertedComp']

    # Fit the model
    model.fit(X_train, y_train)

    # Prepare data for prediction
    missing_data = df[df['ConvertedComp'].isna()]

    # Check if there are missing rows to predict
    if not missing_data.empty:
        X_missing = missing_data[features]

        # Predict missing values
        predicted_values = model.predict(X_missing)

        # Impute missing values
        df.loc[df['ConvertedComp'].isna(), 'ConvertedComp'] = predicted_values

    # Check if there are any remaining missing values
    missing_count_after = df['ConvertedComp'].isna().sum()
    print(f"Number of missing values in 'ConvertedComp' after imputation: {missing_count_after}")
else:
    print("There are no missing values in 'ConvertedComp' to impute.")
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq".

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


```python
# Function to normalize compensation based on frequency
def normalize_compensation(row):
    if row['CompFreq'] == 'Yearly':
        return row['CompTotal']  # No change needed for yearly compensation
    elif row['CompFreq'] == 'Monthly':
        return row['CompTotal'] * 12  # Convert monthly to annual
    elif row['CompFreq'] == 'Weekly':
        return row['CompTotal'] * 52  # Convert weekly to annual
    else:
        return None  # In case of any other value

# Apply the function to create the new column
df['NormalizedAnnualCompensation'] = df.apply(normalize_compensation, axis=1)

# Display the first few rows to check the new column
print(df[['CompFreq', 'CompTotal', 'NormalizedAnnualCompensation']].head())
```
